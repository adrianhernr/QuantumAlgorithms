This program defines the rotation matrix R using the given angles theta and phi. It then applies the rotation to the initial state vector psi and prints the resulting state vector.

```python
from sympy import symbols, cos, sin
from sympy import *

theta, phi = symbols('theta phi', real=True)

# Define Pauli matrices
X = Matrix([[0, 1], [1, 0]])
Y = Matrix([[0, -I], [I, 0]])
Z = Matrix([[1, 0], [0, -1]])

# Define the rotation matrix
R = cos(theta/2) * eye(2) - I * sin(theta/2) * (cos(phi) * X + sin(phi) * Y)

# Define the initial state vector
psi = Matrix([[1], [0]])

# Perform the rotation
psi_rotated = R * psi

# Print the resulting state vector
print(psi_rotated)
```
