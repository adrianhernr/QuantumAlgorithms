```
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, Aer, execute

q = QuantumRegister(3)
c = ClassicalRegister(3)
qc = QuantumCircuit(q, c)

initial_state = [0,1]   # Define initial_state as |1>
qc.initialize(initial_state, 2) 

qc.h(q[0])
qc.h(q[1])
qc.h(q[2])
qc.barrier()
qc.cx(q[0], q[2])
qc.cx(q[1], q[2])
qc.h(q[0])
qc.h(q[1])
qc.barrier()
qc.measure(q,c)

backend = Aer.get_backend('qasm_simulator')
result = execute(qc,backend)
out_state = result.result()
print(out_state.get_counts(qc))
qc.draw()
```

**Quantum algorithms over classical**

Deutsch’s algorithm are often called promise problems, whereby the structure of
the solution space is promised to be of some form and by carefully using superposition, entanglement and interference we can extract information about that structure.
The reason these problems obtain exponential improvement over all known classical algorithms is that we calculate every point using quantum parallelism whereas classically one has to calculate every point in the solution space in order to obtain full knowledge about this structure.  

However there are problems in which we want to find the $x$ that gives $f(x)$ with the desired property.

Thus far, quantum computers can only provide square-root improvement to such
query-based problems.
